```python
import pandas as pd
import numpy as np

file = '../../data/network_2017-08-01_2017-10-31__processedat_2017-11-24.csv'
```

```python
df = pd.read_csv(file)
```

```python
df.head()
```

```python
df[(df['language1']=='JavaScript') & (df['language2']=='Python')]['common_actors'].values[0]
```

```python
languages = df['language1'].unique()
```

```python
TOP = 10
languages = (df[df['language1'] == df['language2']]
             .sort_values('common_actors', ascending=False)
             .iloc[0:TOP]['language1']
             .values)
```

```python
len(languages)
```

```python
df.shape
```

```python
m = np.zeros((len(languages), len(languages)))

for i, l1 in enumerate(languages):
    for j, l2 in enumerate(languages):
        if l1==l2:
            m[i, j] = df[(df['language1']==l1) & (df['language2']==l2)]['common_actors'].values[0]
        else:
            m[i, j] = df[((df['language1']==l1) & (df['language2']==l2)) 
                        | ((df['language1']==l2) & (df['language2']==l1))]['common_actors'].values[0]
```

```python
matrix_raw_actors = pd.DataFrame(m, columns=languages, index=languages)
```

```python
matrix_raw_actors.to_csv('../../app/data/experimental_matrix.csv')
```

```python
matrix_raw_actors.to_csv('../../data/experimental_matrix_values.csv', header=False, index=False)
```
